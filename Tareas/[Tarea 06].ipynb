### Serie de Taylor




```python
import sympy as sp
import numpy as np

# 1. Calcular derivada
# Evaluar la derivada en x = 0
# evaluar_en_cero = lambda expr: expr.subs(x, 0)

x = sp.symbols('x')#crear variable simbolica

f = 1/(25 * x**2 + 1)#funcion
f1 = sp.diff(f, x)#derivada
print('1.',f1)#imprimir derivada
f11 = f1.subs(x, 0)
# Imprimir el resultado de la evaluación
print('  Evaluación de la derivada en x = 0:', f11)

f2 = sp.diff(f1, x)
print('2.',f2)
f22 = f2.subs(x,0)
print('Evaluación de la derivada en x = 0:', f22)

f3 = sp.diff(f2, x)
print('3.',f3)
f33 = f3.subs(x,0)
print('Evaluación de la derivada en x = 0:', f33)

f4 = sp.diff(f3, x)
print('4.',f4)
f44 = f4.subs(x,0)
print('Evaluación de la derivada en x = 0:', f44)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def funcion(x):
    return np.arctan(x)

def fa(x):
    return 45 + 1/2*(x-1) - 1/4*(x-1)**2 + 1/12 * (x-1)**3 + 0 -3/120*(x-1)**5

# Crear un rango de valores de x
x = np.linspace(-25, 25, 200)  # Ajusta el rango según lo necesites para evitar problemas con ln(x) y exp(x)
plt.ylim(-20, 75) 
# Calcular los valores de y para cada función
yy_f = fa(x)
gg_f = funcion(x)
# Graficar las funciones
plt.plot(x, gg_f, label=r"$funcion O(x) ")
plt.plot(x, yy_f, label=r"$fa(x)",linestyle = '--')

# Configurar el gráfico
plt.xlabel("x")
plt.ylabel("funcion O(x), fa(x)")
plt.title("Serie de Taylor = ")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

### Polinomio de Lagrange




```python
xs = [-2, 1, 0, 1, 2]
ys = [-1/49, -1/24, 1, 1/26, 1/51]

x = np.linspace(-10, 10, 200)

plt.ylim(-20, 20) 

def f(x: float) -> float:
    L0 = ((x - xs[1]) * (x - xs[2]) * (x - xs[3])* (x - xs[4])) / ((xs[0] - xs[1]) * (xs[0] - xs[2]) * (xs[0] - xs[3])* (xs[0] - xs[4]))
    L1 = ((x - xs[0]) * (x - xs[2]) * (x - xs[3])* (x - xs[4])) / ((xs[1] - xs[0]) * (xs[1] - xs[2]) * (xs[1] - xs[3])* (xs[1] - xs[4]))
    L2 = ((x - xs[0]) * (x - xs[1]) * (x - xs[3])* (x - xs[4])) / ((xs[2] - xs[0]) * (xs[2] - xs[1]) * (xs[2] - xs[3])* (xs[2] - xs[4]))
    L3 = ((x - xs[0]) * (x - xs[1]) * (x - xs[2])* (x - xs[4])) / ((xs[3] - xs[0]) * (xs[3] - xs[1]) * (xs[3] - xs[2])* (xs[3] - xs[4]))
    L4 = ((x - xs[0]) * (x - xs[1]) * (x - xs[2])* (x - xs[3])) / ((xs[4] - xs[0]) * (xs[4] - xs[1]) * (xs[4] - xs[2])* (xs[4] - xs[3]))
    return L0 * ys[0] + L1 * ys[1] + L2 * ys[2] + L3 * ys[3] + L4 * ys[4]


plt.plot(x, [f(val) for val in x], label='Interpolated function')
plt.scatter(xs, ys, color='red', label='Data points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Lagrange Interpolating Polynomial')
plt.legend()
plt.show()
```

### Polinomio de Lagrange



```python
xs = [-2, 1, 0, 1, 2]
ys = [-63.43, -45, 0, 45, 63.43]

x = np.linspace(-20, 20, 200)

plt.ylim(-50, 50) 

def f(x: float) -> float:
    L0 = ((x - xs[1]) * (x - xs[2]) * (x - xs[3])* (x - xs[4])) / ((xs[0] - xs[1]) * (xs[0] - xs[2]) * (xs[0] - xs[3])* (xs[0] - xs[4]))
    L1 = ((x - xs[0]) * (x - xs[2]) * (x - xs[3])* (x - xs[4])) / ((xs[1] - xs[0]) * (xs[1] - xs[2]) * (xs[1] - xs[3])* (xs[1] - xs[4]))
    L2 = ((x - xs[0]) * (x - xs[1]) * (x - xs[3])* (x - xs[4])) / ((xs[2] - xs[0]) * (xs[2] - xs[1]) * (xs[2] - xs[3])* (xs[2] - xs[4]))
    L3 = ((x - xs[0]) * (x - xs[1]) * (x - xs[2])* (x - xs[4])) / ((xs[3] - xs[0]) * (xs[3] - xs[1]) * (xs[3] - xs[2])* (xs[3] - xs[4]))
    L4 = ((x - xs[0]) * (x - xs[1]) * (x - xs[2])* (x - xs[3])) / ((xs[4] - xs[0]) * (xs[4] - xs[1]) * (xs[4] - xs[2])* (xs[4] - xs[3]))
    return L0 * ys[0] + L1 * ys[1] + L2 * ys[2] + L3 * ys[3] + L4 * ys[4]


plt.plot(x, [f(val) for val in x], label='Interpolated function')
plt.scatter(xs, ys, color='red', label='Data points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Lagrange Interpolating Polynomial')
plt.legend()
plt.show()
```
