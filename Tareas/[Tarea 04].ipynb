```python
import math

# Constantes
L = 10  # cm Logitud
r = 1   # cm radio
V = 12.4  # Volumen
tolerancia = 1e-2
# Función que calcula  altura h del agua
def calcularVolumen(h):
    return L * (0.5 * math.pi * r**2 - r**2 * math.asin(h / r) - h * math.sqrt(r**2 - h**2)) - V

# Método de bisección para encontrar la raíz de la función (profundidad del agua)
def biseccionVolumen(f, a, b, tolerancia):
    while (b - a) / 2 > tolerancia:
        m = (a + b) / 2
        if f(m) == 0:
            return m
        elif f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2

# Intervalo inicial y llamada a la función de bisección
a = 0
b = r
profundidad = biseccionVolumen(calculoVolumen, a, b, tolerancia)

# Imprimir el resultado
print(f"La profundidad es {profundidad:.3f}")
```

```python
import math

# Constantes
S = 300  # Altura m
m = 0.25  #kg
g = 9.81  # g en m/s^2
k = 0.1   # Constante de resistencia del aire Ns/m
tolerancia=10e-2
# Función que calcula la altura en función del tiempo t
def Altura(t):
    return S - (m * g / k) * t + (m**2 * g / k**2) * (1 - math.exp(-k * t / m))

# Método de bisección para encontrar el tiempo en que la altura es 0
def biseccionTiempo(f, a, b, tolerancia):
    while (b - a) / 2 > tolerancia:
        m = (a + b) / 2
        if f(m) == 0:
            return m
        elif f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2

# Intervalo inicial y llamada a la función de bisección
a = 0
b = 100
tiempo_tardado = biseccionTiempo(Altura, a, b,tolerancia)

# Imprimir el resultado
print(f"El tiempo  es {tiempo_tardado} s")
```

```python
def bisecion(a=1, b=2, equation=lambda x: x**3 - x - 1, tol=10**(-4), N=20):
    for i in range(N):
        m = (a + b) / 2
        if equation(m) == 0 or (b - a) / 2 < tol:
            return m, i
        elif equation(a) * equation(m) < 0:
            b = m
        else:
            a = m
    return m, N

# Llamada a la función bisección
r, i = bisecion(a=1, b=2, equation=lambda x: x**3 - x - 1, tol=10**(-4), N=20)

# Imprimir el resultado
print(f"La bisección es {r} en {i} iteraciones")
```
