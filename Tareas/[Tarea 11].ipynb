### 4. Utilice el método de Gauss-Siedel para resolver los sistemas lineales en el ejercicio 1, con TOL = 10^-3.

```python
def metodo_gauss_seidel(A, b, x0, tol=1e-3, num_iterations=100):
    n = len(b)
    x = x0.copy()
    for iteration in range(1, num_iterations + 1):
        for i in range(n):
            sum_ax = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x[i] = (b[i] - sum_ax) / A[i][i]  

        # Criterio de convergencia
        if np.linalg.norm(np.dot(A, x) - b, ord=np.inf) < tol:
            print(f"Convergencia alcanzada en la iteración {iteration}: {x}")
            return x
        print(f"Iteración {iteration}: {x}")
    return x

```
