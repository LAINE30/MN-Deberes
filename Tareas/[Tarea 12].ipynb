### Ejercicio A

```python


def ODE_euler(a: float, b: float, f, y_t0: float, N: int):
    h = (b - a) / N  # tamaño del paso
    t_values = [a + i*h for i in range(N+1)]  # valores de t
    y_values = [y_t0]  # valores de y
    y = y_t0
    
    for i in range(1, N+1):
        y = y + h * f(t_values[i-1], y)  # Método de Euler
        y_values.append(y)
        
    # Imprimir la tabla de resultados
    print("Resultados del Método de Euler (Literal a)")
    print("=========================================")
    print(f"{'t_i':<10} {'y(t_i)':<10}")
    for t, y in zip(t_values, y_values):
        print(f"{t:<10.2f} {y:.6f}")    
    return y_values, t_values, h

# Parámetros para el problema a
a = 1
b = 2
y_t0_a = 1
h_a = 0.1
N_a = int((b - a) / h_a)

# Definir la función f para el problema a
def f_a(t, y):
    return y / t - (y / t)**2

ys_a, ts_a, h_a = ODE_euler(a=a, b=b, f=f_a, y_t0=y_t0_a, N=N_a)

```

```python
# Calcular el error real para el literal a
error_a = [abs(y_real_a(t) - y_approx) for t, y_approx in zip(ts_a, ys_a)]

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(ts_a, ys_a, marker='o', linestyle='-', color='c', label='Aproximación de Euler')
plt.plot(ts_a, [y_real_a(t) for t in ts_a], linestyle='--', label='Solución Real', color='brown')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Aproximación de Euler y Solución Real (Literal a)')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir los errores reales
print("Errores reales (Literal a)")
for t, err in zip(ts_a, error_a):
    print(f"t = {t:.2f}, Error Real = {err:.6f}")
```
