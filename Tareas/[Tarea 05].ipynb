```python
#METODO DE NEWTON-RAPHSON
import numpy as np


def newton_raphson(f, f_prime, x0, tol=1e-10, max_iter=100):
    """
    Método de Newton-Raphson para encontrar una raíz de la función f(x).
    
    Parámetros:
    f : función -> La función cuya raíz queremos encontrar.
    f_prime : función -> La derivada de f(x).
    x0 : float -> Punto inicial para la iteración.
    tol : float -> Tolerancia para la convergencia.
    max_iter : int -> Máximo número de iteraciones permitidas.
    
    Retorna:
    Una aproximación de la raíz.
    """
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        
        if abs(fpx) < 1e-10:  # Evita divisiones por números muy pequeños
            raise ValueError("La derivada es muy pequeña; el método no converge.")
        
        x_new = x - fx / fpx  # Fórmula del método de Newton
        
        # Imprime información de las iteraciones (opcional)
        print(f"Iteración {i+1}: x = {x_new}, f(x) = {f(x_new)}")
        
        # Verifica si la solución es suficientemente precisa
        if abs(x_new - x) < tol:
            return x_new
        
        x = x_new  # Actualiza el valor de x
    
    raise ValueError("El método no convergió en el número máximo de iteraciones.")

# Ejemplo de uso
if __name__ == "__main__":
    # Define la función f(x) y su derivada f'(x)
    def f(x):
        return -x**3 - np.cos(x) # Queremos encontrar sqrt(2)
    
    def f_prime(x):
        return -3*x**2 + np.sin(x)  # Derivada de f(x)
    
    # Llamamos al método de Newton con un valor inicial
    raiz = newton_raphson(f, f_prime, x0=-1)
    print(f"La raíz encontrada es: {raiz}")
```
