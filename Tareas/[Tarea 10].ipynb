```python
def multiplicar_matrices(A, B):
    # Verificar si el número de columnas de A es igual al número de filas de B
    if len(A[0]) != len(B):
        raise ValueError("No se pueden multiplicar: las dimensiones no son compatibles.")
    
    # Inicializar la matriz resultado con ceros
    resultado = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

    # Multiplicación de matrices
    for i in range(len(A)):            # Recorre filas de A
        for j in range(len(B[0])):     # Recorre columnas de B
            for k in range(len(B)):    # Recorre filas de B (o columnas de A)
                resultado[i][j] += A[i][k] * B[k][j]

    return resultado
```

```python
# Función para calcular el determinante de una matriz n x n
def determinante(matriz):
    n = len(matriz)
    # Caso base: matriz 1x1
    if n == 1:
        return matriz[0][0]
    # Caso base: matriz 2x2
    if n == 2:
        return matriz[0][0] * matriz[1][1] - matriz[0][1] * matriz[1][0]
    
    # Expansión de Laplace para matrices mayores
    det = 0
    for col in range(n):
        sub_matriz = [fila[:col] + fila[col+1:] for fila in matriz[1:]]
        signo = (-1) ** col
        det += signo * matriz[0][col] * determinante(sub_matriz)
    return det

# Función para calcular la matriz de cofactores
def matriz_cofactores(matriz):
    n = len(matriz)
    cofactores = []
    for i in range(n):
        fila_cofactores = []
        for j in range(n):
            sub_matriz = [fila[:j] + fila[j+1:] for k, fila in enumerate(matriz) if k != i]
            signo = (-1) ** (i + j)
            fila_cofactores.append(signo * determinante(sub_matriz))
        cofactores.append(fila_cofactores)
    return cofactores

# Función para transponer una matriz
def transpuesta(matriz):
    return [list(fila) for fila in zip(*matriz)]

# Función para calcular la inversa de una matriz
def inversa(matriz):
    det = determinante(matriz)
    if det == 0:
        return None  # Matriz singular, no tiene inversa
    
    cofactores = matriz_cofactores(matriz)
    adjunta = transpuesta(cofactores)
    inversa = [[adjunta[i][j] / det for j in range(len(matriz))] for i in range(len(matriz))]
    return inversa

# Función principal para evaluar la matriz
def evaluar_matriz(matriz):
    det = determinante(matriz)
    print("Matriz:")
    for fila in matriz:
        print(fila)
    print(f"Determinante = {det}")
    
    if det != 0:
        inv = inversa(matriz)
        print("Es no singular. Su inversa es:")
        for fila in inv:
            print(fila)
    else:
        print("Es singular. No tiene inversa.")
    print("-" * 30)

```

```python
# Definimos la matriz de coeficientes A
A = [
    [1, -1, 2, -1],
    [1, 0, -1, 1],
    [2, 1, 3, -4],
    [0, -1, 1, -1]
]

# Vectores de constantes para los dos sistemas
B1 = [6, 4, -2, 5]  # Primer sistema
B2 = [1, 1, 2, -1]  # Segundo sistema
```

```python

# Función para calcular el determinante de una matriz
def determinante(matriz):
    if len(matriz) == 2:
        return matriz[0][0]*matriz[1][1] - matriz[0][1]*matriz[1][0]

    det = 0
    for c in range(len(matriz)):
        submatriz = [fila[:c] + fila[c+1:] for fila in (matriz[1:])]
        det += ((-1)**c) * matriz[0][c] * determinante(submatriz)
    return det

# Función para encontrar la matriz inversa
def inversa(matriz):
    det = determinante(matriz)
    if det == 0:
        return None  # No tiene inversa

    # Matriz de cofactores
    cofactores = []
    for r in range(len(matriz)):
        fila_cofactores = []
        for c in range(len(matriz)):
            menor = [fila[:c] + fila[c+1:] for i, fila in enumerate(matriz) if i != r]
            fila_cofactores.append(((-1) ** (r + c)) * determinante(menor))
        cofactores.append(fila_cofactores)

    # Transponer y dividir por el determinante
    cofactores_T = list(map(list, zip(*cofactores)))
    inversa = [[elem / det for elem in fila] for fila in cofactores_T]
    return inversa

# Función para multiplicar matrices
def multiplicar_matrices(A, B):
    resultado = []
    for i in range(len(A)):
        suma = 0
        for j in range(len(B)):
            suma += A[i][j] * B[j]
        resultado.append(suma)
    return resultado

# Cálculo del determinante y la inversa de A
det_A = determinante(A)
print(f"Determinante de A: {det_A}")

if det_A != 0:
    A_inv = inversa(A)
    print("Matriz inversa de A:")
    for fila in A_inv:
        print(fila)

    # Resolver ambos sistemas
    X1 = multiplicar_matrices(A_inv, B1)
    X2 = multiplicar_matrices(A_inv, B2)

    print("\nSolución del primer sistema:", X1)
    print("Solución del segundo sistema:", X2)
else:
    print("La matriz A es singular y no tiene inversa.")
```

```python
from sympy import symbols, Eq, solve

# Definir la variable simbólica
alpha = symbols('alpha')

# Definir la matriz A
A = [
    [1, -1, alpha],
    [2, 2, 1],
    [0, alpha, -3/2]
]

# Función para calcular el determinante de una matriz 3x3
def determinante_3x3(m):
    return (
        m[0][0] * (m[1][1] * m[2][2] - m[1][2] * m[2][1]) -
        m[0][1] * (m[1][0] * m[2][2] - m[1][2] * m[2][0]) +
        m[0][2] * (m[1][0] * m[2][1] - m[1][1] * m[2][0])
    )

# Calcular el determinante de A
det_A = determinante_3x3(A)

# Resolver la ecuación det(A) = 0 para encontrar los valores de alpha
valores_singulares = solve(Eq(det_A, 0), alpha)

# Imprimir resultados
print("Determinante de A:", det_A)
print("Valores de alpha que hacen la matriz singular:", valores_singulares)
```

```python
def gauss_elimination(AB, b):
    n = len(AB)

    # Formar la matriz aumentada
    for i in range(n):
        AB[i].append(b[i])

    # Eliminación hacia adelante
    for i in range(n):
        # Pivoteo parcial
        max_row = max(range(i, n), key=lambda r: abs(A[r][i]))
        AB[i], AB[max_row] = AB[max_row], AB[i]

        # Hacer el pivote igual a 1
        pivot = AB[i][i]
        if pivot == 0:
            raise ValueError("El sistema no tiene solución única")
        B[i] = [elem / pivot for elem in AB[i]]

        # Hacer ceros debajo del pivote
        for j in range(i + 1, n):
            factor = AB[j][i]
            AB[j] = [AB[j][k] - factor * AB[i][k] for k in range(n + 1)]

    # Sustitución hacia atrás
    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = AB[i][-1] - sum(A[i][j] * x[j] for j in range(i + 1, n))

    return x
```

```python
import numpy as np
import logging

def descomposicion_LU(A: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """Realiza la descomposición LU de una matriz cuadrada A.
    [IMPORTANTE] No se realiza pivoteo.

    ## Parameters

    ``A``: matriz cuadrada de tamaño n-by-n.

    ## Return

    ``L``: matriz triangular inferior con unos en la diagonal.

    ``U``: matriz triangular superior. Se obtiene de la matriz ``A`` después de aplicar la eliminación gaussiana.
    """

    A = np.array(
        A, dtype=float
    )  # convertir en float, porque si no, puede convertir como entero

    assert A.shape[0] == A.shape[1], "La matriz A debe ser cuadrada."
    n = A.shape[0]

    L = np.zeros((n, n), dtype=float)

    for i in range(0, n):  # loop por columna

        # --- determinar pivote
        if A[i, i] == 0:
            raise ValueError("No existe solución única.")

        # --- Eliminación: loop por fila
        L[i, i] = 1  # La diagonal de L es 1
        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]  # Multiplicador
            A[j, i:] = A[j, i:] - m * A[i, i:]  # Actualizar la fila j

            L[j, i] = m  # Guardar el multiplicador en L

        logging.info(f"\n{A}")

    if A[n - 1, n - 1] == 0:
        raise ValueError("No existe solución única.")

    return L, A

```

```python
import numpy as np

def lu_decomposition(A):
    n = len(A)
    L = np.eye(n)
    U = A.astype(float)
    
    for i in range(n):
        for j in range(i+1, n):
            factor = U[j, i] / U[i, i]
            L[j, i] = factor
            U[j, i:] -= factor * U[i, i:]
    
    return L, U

def forward_substitution(L, b):
    n = len(L)
    y = np.zeros(n)
    for i in range(n):
        y[i] = b[i] - np.dot(L[i, :i], y[:i])
    return y

def backward_substitution(U, y):
    n = len(U)
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i+1:], x[i+1:])) / U[i, i]
    return x

def solve_using_lu(A, b):
    L, U = lu_decomposition(A)
    y = forward_substitution(L, b)
    x = backward_substitution(U, y)
    return x

# Sistemas de ecuaciones
systems = [
    {
        "A": np.array([[2, -1, 1],
                        [3, 3, 9],
                        [3, 3, 5]]),
        "b": np.array([-1,0,4])
    },
    {
        "A": np.array([[1.012, -2.132, 3.104],
                        [-2.132, 4.096, -7.013],
                        [3.104, -7.013, 0.014]]),
        "b": np.array([1.984, -5.049, -3.895])
    },
    {
        "A": np.array([[2, 0, 0, 0],
                        [1, 1.5, 0, 0],
                        [0, -3, 0.5, 0],
                        [2,-2,1,1]]),
        "b": np.array([3, 4.5, -6.6, 0.8])
    },
    {
        "A": np.array([[2.1756, 4.0231, -2.1732, 5.1967],
                        [-4.0231, 6.0000, 0, 1.1973],
                        [-1.0, -5.2107, 1.1111, 0],
                        [6.0235, 7.0, 0, -4.1561]]),
        "b": np.array([17.102, -6.1593, 3.0004, 0.0])
    }
]

# Resolver los sistemas
solutions = [solve_using_lu(system["A"], system["b"]) for system in systems]

for i, sol in enumerate(solutions):
    print(f"Solución del sistema {i+1}: {sol}")
```
